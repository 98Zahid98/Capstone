```python
import numpy as np
import pandas as pd
import plotly.graph_objs as go
import plotly.express as px

# Example DataFrame (df) containing 'ExitRates' and 'BounceRates'
# Assuming df already exists with 'ExitRates' and 'BounceRates'

# Define x and y from the DataFrame
x = df['ExitRates']
y = df['BounceRates']

# Calculate correlation coefficient
result = np.corrcoef(x, y)
print(result)

# Linear regression -> m is the slope, b is the y-intercept
m, b = np.polyfit(x, y, deg=1)

# Create scatter plot and regression line using Plotly
scatter = go.Scatter(
    x=x, 
    y=y, 
    mode='markers', 
    marker=dict(color='green', opacity=0.5), 
    name='Data Points'
)

# Create the regression line
line = go.Scatter(
    x=x, 
    y=m*x + b, 
    mode='lines', 
    line=dict(color='red'), 
    name='Regression Line'
)

# Layout for the interactive plot
layout = go.Layout(
    title='Exit Rates vs Bounce Rates',
    xaxis=dict(title='ExitRates'),
    yaxis=dict(title='BounceRates'),
    showlegend=True
)

# Combine scatter plot and regression line
fig = go.Figure(data=[scatter, line], layout=layout)

# Show interactive plot
fig.show()
```
