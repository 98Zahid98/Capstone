# <font color='purple'> Decision Tree (with K-fold cross validation) </font> 

```python
X = df2.drop('Returning_Type', axis=1)
y = df2['Returning_Type']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X ,y, 
                                                    test_size = 0.2, 
                                                    shuffle=True, 
                                                    stratify=y,
                                                    random_state=SEED)
```

```python
# Training the model

# Import the Decision Tree algorithm
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier(random_state=SEED)
```

```python
# Hyperparameter tuning using K-fold cross validation
# ... via Grid Search method
from sklearn.model_selection import GridSearchCV

param_grid = {'criterion': ['gini', 'entropy'], 
              'max_depth': range(1,10),
              'min_samples_split': range(1, 10),
              'min_samples_leaf': range(1, 5)}

gs_clf = GridSearchCV(clf,
                      param_grid, 
                      cv=5,
                      scoring='f1_macro',
                      n_jobs=-1)
gs_clf.fit(X_train, y_train)
```

```python
# These are the best parameters for the decision tree classifier
gs_clf.best_estimator_
```

```python
# Training the final model using the best parameters from above fine-tuning
classifier = gs_clf.best_estimator_
classifier.fit(X_train, y_train)
```

```python
# Visualize the tree
from sklearn import tree

plt.figure(figsize=(40,20))
dtree_plot = tree.plot_tree(classifier,
                           feature_names = df2.columns,
                           class_names = ['0','1'],
                           filled = True,
                           fontsize = 14)
plt.show()
```

```python
# Testing the model

# Kept aside some data to test - X_test
y_pred = classifier.predict(X_test)

compare_df2 = pd.DataFrame({"Desired Output (Actuals)": y_test, 
                           "Predicted Output": y_pred})
```

```python
compare_df2[:10]
```

```python
# Evaluating the model

# Classification Report
print("Classification report:")
print()
print(classification_report(y_test, y_pred, target_names=['0','1']))
print()

# Confusion Matrix
print("Confusion Matrix:")
confusion_matrix(y_test, y_pred)
```

# <font color='purple'> SMOTE (Synthetic Minority Oversampling Technique)</font> 

### Oversampling technique that creates synthetic minority class data points using K Nearest Neighbors to balance the dataset.

```python
X = df4.drop('Returning_Type', axis=1)
y = df4['Returning_Type']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X ,y, test_size = 0.2, random_state=SEED)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
model.fit(X_train, y_train)
y_predict = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy score before SMOTE:", accuracy_score(y_test, y_predict))
pd.crosstab(y_test, y_predict)
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE()
```

```python
# SMOTE is only applied to the training data because you are using SMOTE to gain
# an improvement in operational performance

X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)
```

```python
from collections import Counter
print("Before SMOTE:", Counter(y_train))
print("After SMOTE:", Counter(y_train_smote))
```

```python
model.fit(X_train_smote, y_train_smote)
y_predict = model.predict(X_test)
print("Accuracy score after SMOTE:", accuracy_score(y_test, y_predict))
pd.crosstab(y_test, y_predict)
```

```python
# Evaluating the model

# Classification Report
print("Classification report:")
print()
print(classification_report(y_test, y_predict, target_names=['0','1']))
print()

# Confusion Matrix
print("Confusion Matrix:")
confusion_matrix(y_test, y_predict)
```
